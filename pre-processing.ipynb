```python
import pandas as pd
import numpy as np
import datetime
```

__Data Import:__

```python
data = pd.read_csv('hq_bookings.csv')
```

```python
data.head()
```

```python
print ('Number of rows before removing the Unconfirmed row:', len(data))
```

__Platform column simplified:__

```python
data = data.replace({'Platform 1': 1, 'Platform 2': 2, 'Platform 3':3})
```

__one row out of 126K rows had an Unconfirmed status, removed:__

```python
data = data[data.Order_Status != 'Unconfirmed']
```

```python
print ('Number of rows before After the Unconfirmed row:', len(data))
```

__ReIndexing:__

```python
data.index = range(len(data))
```

```python
data.head()
```

__Adding  rows of Booking_Month and Booking_Day:__

```python
booking_zeros =  pd.DataFrame(np.zeros((len(data), 1)))
data.insert(loc=1, column='Booking_Day', value=booking_zeros)
data.insert(loc=1, column='Booking_Month', value=booking_zeros)
```

__Cutting the value of day from the date:__

```python
data['Booking_Day'] = data['Booking_Date'].apply(lambda st: st[st.find("/")+1:st.rfind("/")])
```

__Cutting the value of month from the date:__

```python
data['Booking_Month'] = [x.split('/')[0] for x in data['Booking_Date']]
```

```python
data.tail()
```

__Adding  rows of Checkin_Month and Checkin_Day:__

```python
checkin_zeros =  pd.DataFrame(np.zeros((len(data), 1)))
data.insert(loc=5, column='Checkin_Day', value=booking_zeros)
data.insert(loc=5, column='Checkin_Month', value=booking_zeros)
```

```python
data.head()
```

__Cutting the value of day from the date:__

```python
data['Checkin_Day'] = data['Check_In_Date'].apply(lambda st: st[st.find("/")+1:st.rfind("/")])
```

__Cutting the value of month from the date:__

```python
data['Checkin_Month'] = [x.split('/')[0] for x in data['Check_In_Date']]
```

```python
data.tail()
```

__Adding a new column for total_days:__

```python
checkin_zeros =  pd.DataFrame(np.zeros((len(data), 1)))
data.insert(loc=7, column='Total_Days', value=booking_zeros)
```

```python
data.tail()
```

__Total days calculated by order amount, times length of each stay:__

```python
data['Total_Days'] = data.Order_Amount * data.Length_of_Stay_In_Nights
```

```python
data.head()
```

__ Countries ranking based on the summation of days all people from the same country booked in all 4 months:__

```python
rank = data.groupby('User_Country_Name')['Total_Days'].agg({'Total_Days':'sum'})
rank = rank.sort_values(by=['Total_Days'], ascending= False).reset_index()
rank
```

__For this is just a test, we keep the first 20 countries only and work on them:__ 

```python
rank = rank.head(20)
rank
```

__New dataset with top 20 countries:__

```python
top_countries = rank['User_Country_Name'].tolist()
new_data = data.loc[data['User_Country_Name'].isin(top_countries)]
print('Top countries are kept which are', new_data.size, 'rows out of', data.size, 'rows.')
```

```python
new_data.describe()
```

__Some countries have very high cancellation rates. The table below shows countries ranked based on the highest churn rate (lowest percentage):__

```python
rank_active_vs_cancelled = new_data.pivot_table(index='Order_Status',columns='User_Country_Name',values='Total_Days').transpose()
rank_active_vs_cancelled =  rank_active_vs_cancelled.fillna(0).astype(int)
rank_active_vs_cancelled['Sum'] = rank_active_vs_cancelled.Active + rank_active_vs_cancelled.Cancelled
rank_active_vs_cancelled['Performance'] = ((rank_active_vs_cancelled['Active'] / rank_active_vs_cancelled['Sum']) * 100).astype(int)
rank_active_vs_cancelled = rank_active_vs_cancelled.sort_values(by=['Performance']).reset_index()
rank_active_vs_cancelled
```

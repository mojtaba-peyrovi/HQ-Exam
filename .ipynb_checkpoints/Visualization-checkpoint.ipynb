```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams["patch.force_edgecolor"] = True
plt.ylim(0, 200)
plt.xlim(0, 9000)
```

```python
data = pd.read_csv('pre-processed-for-visualization.csv')
```

```python
data.head()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,8))
Platform_popularity_order_amount = sns.barplot(x='Platform',y='Order_Amount',data=data)
```

```python

plt.figure(figsize=(12,8))
Platform_popularity_order_count= sns.barplot(x='Platform',y='Order_Amount',data=data)
```

```python
plt.figure(figsize=(12,8))
Status_platform = sns.barplot(x='Platform',y='Total_Days',data=data, hue="Order_Status")
```

```python
top_deals = pd.read_csv('top_deals.csv')
top_deals.head()
```

```python

plt.figure(figsize=(18,8))
user_country = sns.barplot(x='User_Country_Name',y='Total_Days',data=top_deals.head(10))
```

```python
booking_country_month= top_deals.pivot_table(index='Booking_Month',columns='User_Country_Name',values='Total_Days',fill_value=0).astype(int)
```

```python
booking_country_month
```

```python
plt.figure(figsize=(22,18))
sns.heatmap(booking_country_month,cmap='coolwarm',fmt='g')
```
